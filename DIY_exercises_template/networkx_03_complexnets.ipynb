```python
import networkx as nx
```

# Complete Graph

```python
nodes = 10

G = nx.complete_graph(nodes)

layout = nx.spring_layout(G)

nx.draw(G, pos=layout)
```

# Random Graph

```python
edge_probability = 0.5

ER = nx.erdos_renyi_graph(nodes, edge_probability, seed=123)

layout = nx.spring_layout(ER)

nx.draw(ER, pos=layout)
```

```python
print('Nodes:')
print(ER.nodes)

print('')

print('Edges:')
print(ER.edges)
```

# Watts-Strogatz Graph


```python
nodes = 60
neighbours = 4
alpha = 0.9

ws = nx.watts_strogatz_graph(nodes, neighbours, alpha)

layout = nx.spring_layout(ws)

nx.draw(ws, 
        node_color='k', 
        node_size=500, 
        node_shape='h',
        edge_color='r', 
        width=1)
```
