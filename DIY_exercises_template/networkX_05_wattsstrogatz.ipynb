```python
import networkx as nx
```

```python
nodes      = 10
neighbours = 2
alpha      = 0.1

ws = nx.watts_strogatz_graph(nodes, neighbours, alpha, seed=11111)

layout = nx.spectral_layout(ws)

nx.draw(ws, node_color='red')
```

```python


nx.draw_networkx_nodes(ws, layout, 
                       alpha=1, 
                       node_size=100, 
                       node_color='cornflowerblue', 
                       node_shape='o')

nx.draw_networkx_edges(ws, layout, 
                       alpha=.4, 
                       width=3, 
                       edge_color='tomato');
```
