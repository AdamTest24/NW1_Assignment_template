```python
import networkx as nx
```

# Create Graph

```python
graph = nx.Graph()

graph.add_node('Node A')
graph.add_node('Node B')
graph.add_node('Node C')

graph.add_edge('Node A', 'Node C')
graph.add_edge('Node A', 'Node B')
graph.add_edge('Node B', 'Node C')
```

# Plot Graph


```python
from numpy import array

fig = nx.draw(graph)
```

# Plot Graph with Layout


```python
layout = nx.spectral_layout(graph)

fig = nx.draw(graph, layout, 
              node_size=[1000, 1000, 1000],
              node_color='red')
```

More options:
https://networkx.org/documentation/stable/reference/drawing.html#module-networkx.drawing.layout


# Add Labels

```python
layout = nx.spectral_layout(graph)

fig = nx.draw(graph, layout, 
              node_size = [1000, 1000, 1000])

labels = {
    # key   : value
    'Node A': 'A',
    'Node B': 'B',
    'Node C': 'C',
}

nx.draw_networkx_labels(graph, layout, labels=labels);

```
